```python
data = np.genfromtxt('iris.data.txt', delimiter=',', usecols=(0,1,2,3))
```

```python
data
```

```python
len(data)
```

```python
#valores da primeira coluna
data[:,0]
```

```python
# valores das  cepalas da ires cesosa 50 primeiras
compSepalaSetosa=data[:50,0]
```

```python
plt.plot(compSepalaSetosa,c='red', ls=':',marker='s',ms=8)
```

```python
compSepalaVerscolor=data[50:100,0]
```

```python
plt.plot(compSepalaVerscolor,c='blue', ls=':',marker='s',ms=8)
```

```python
plt.plot(compSepalaSetosa,c='red', ls=':',marker='s',ms=8 ,label='sepala setosa')
plt.plot(compSepalaVerscolor,c='blue', ls=':',marker='o',ms=8,label='sepala versicolor')
plt.legend()
plt.show()
```

```python
compSepalaVirginica = data[100:150,0]
plt.plot(compSepalaVirginica,c='green', ls=':',marker='s',ms=8)
```

```python
plt.plot(compSepalaSetosa,c='red', ls=':',marker='s',ms=8 ,label='sepala setosa')
plt.plot(compSepalaVerscolor,c='blue', ls=':',marker='o',ms=8,label='sepala versicolor')
plt.plot(compSepalaVirginica,c='green', ls=':',marker='s',ms=8,label='sepala virginica')
plt.legend()
plt.show()
```
