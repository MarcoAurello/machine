```python
def categoriza_volume_chuva(s):
    if s >= 0 and s <=1.5:
       return 'Media Baixissima'
    elif s > 1.5 and s <=3:
       return 'Media Razoavel'
    elif s > 3 and s <=5:
       return 'Media Normal'
    elif s > 5 and s <=5.5:
       return 'Media acima do esperado'
    elif s > 5.5:
       return 'Media muito acima do esperado'   
```
