```python
df = pd.read_csv('acidentes-2016.csv', delimiter=';' )
```

```python
df['tipo de ocorrencia'].unique()
```

```python
df['data de abertura'] = pd.to_datetime(df['data de abertura'], format='%d/%m/%Y' , errors='ignore')
```

```python
df['hora de abertura '] = pd.to_datetime(df['hora de abertura'], format='%H:%M' , errors='ignore')
```

```python
df.dtypes
```

```python
df[(df['fraction_votes'] ==1) & (df['candidate']== 'Hillary Clinton')]
```

```python
  ocorrenciaMadrugada =  df[(df['hora de abertura '] > '00:00') & (df['hora de abertura '] < '05:00')]
```

```python
ocorenciaJan =  df[(df['data de abertura'] > '2016-01-01') & (df['data de abertura'] < '2016-01-31')]
ocorenciaFev =  df[(df['data de abertura'] > '2016-02-01') & (df['data de abertura'] < '2016-02-29')]
ocorenciaMar =  df[(df['data de abertura'] > '2016-03-01') & (df['data de abertura'] < '2016-03-31')]
ocorenciaAbr =  df[(df['data de abertura'] > '2016-04-01') & (df['data de abertura'] < '2016-04-30')]
ocorenciaMai =  df[(df['data de abertura'] > '2016-05-01') & (df['data de abertura'] < '2016-05-31')]
ocorenciaJun =  df[(df['data de abertura'] > '2016-06-01') & (df['data de abertura'] < '2016-06-30')]
ocorenciaJul =  df[(df['data de abertura'] > '2016-07-01') & (df['data de abertura'] < '2016-07-31')]
```

```python
print("Acidentes em Janeiro: ", len(ocorenciaJan))
print("Acidentes em Fevereiro: ", len(ocorenciaFev)) 
print("Acidentes em Março: ", len(ocorenciaMar)) 
print("Acidentes em Abril: ", len(ocorenciaAbr))
print("Acidentes em Maio: ", len(ocorenciaMai))
print("Acidentes em Junho: ", len(ocorenciaJun)) 
print("Acidentes em Julho: ",len(ocorenciaJul))
```
