```python
Populacao2000 = populacao_pe['Total da população 2000'].sum()
Populacao2010 = populacao_pe['Total da população 2010'].sum()

homens = populacao_pe['Total de homens'].sum()
mulheres= populacao_pe['Total de mulheres'].sum()
rural = populacao_pe['Total da população rural'].sum()
urbano = populacao_pe['Total da população urbana'].sum()  
```

```python
print('população anos 2000: ',Populacao2000 )
print('população anos 2010: ',Populacao2010 )
print('diferença: ', Populacao2010 - Populacao2000)

print('_________________')
print('homens: ',homens)
print('mulheres: ',mulheres)
print('rural: ',rural )
print('urbano: ',urbano )

```
