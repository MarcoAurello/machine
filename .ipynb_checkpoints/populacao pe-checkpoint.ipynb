```python
homens =populacao_pe['Total de homens'].sum()
mulheres =populacao_pe['Total de mulheres'].sum()
```

```python
print('Quantidade de homens: ',homens)
print('Quantidade de mulher: ',mulheres)
```

```python
total = homens + mulheres
diferença = mulheres - homens
```

```python
def plot_grafico():
    labels ='Homens', 'Mulheres'
    populacao = [homens,mulheres]
    cores=['green','red']
    explode=(0.1,0)
    total = sum(populacao)
    plt.pie(populacao,explode=explode,labels = labels ,colors=cores,
           autopct= lambda p: '{:.0f}'.format(p* total / 100), shadow = True,
           startangle =90)
    plt.axis('equal')
    plt.show()
```

```python
plot_grafico()
```
