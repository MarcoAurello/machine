```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario,c='red',ls='--',marker='s',ms='10')
plt.show()
```

```python
salarios_marco = np.array([100,200,400,300])
salarios_ana = np.array([300,200,300,400])

```

```python
plt.plot(salarios_marco,c = 'black',marker='s',ls='--',ms=8,label="salarios marco")
plt.plot(salarios_ana,ls='--',marker='s',c = 'red',ms=8,label="salarios ana")
plt.show()
```

```python
#legenda
plt.plot(salarios_marco,c = 'black',marker='s',ls='--',ms=8,label="salarios marco")
plt.plot(salarios_ana,ls='--',marker='s',c = 'red',ms=8,label="salarios ana")
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marco,c = 'black',marker='s',ls='--',ms=8,label="salarios marco")
plt.plot(salarios_ana,ls='--',marker='^',c = 'red',ms=8,label="salarios ana")
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marco,c = 'black',marker='s',ls='--',ms=8,label="salarios marco")
plt.plot(salarios_ana,ls='--',marker='^',c = 'red',ms=8,label="salarios ana")
plt.legend(loc="upper right")
plt.show()
```

```python
# mais uma linha
salario_pedro = [200,100,270,160]
plt.plot(salario_pedro,c = 'blue',ls='--',marker='o',label='salario Pedro')
plt.plot(salarios_marco,c = 'black',marker='s',ls='--',ms=8,label="salarios Marco")
plt.plot(salarios_ana,ls='--',marker='^',c = 'red',ms=8,label="salarios Ana")
plt.legend(loc="upper left")
plt.show()
```
